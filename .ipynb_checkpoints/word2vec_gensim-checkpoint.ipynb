```python
from gensim.corpora import WikiCorpus
from zemberek import TurkishTokenizer
from gensim import utils
```

```python
def tknz(inp, token_min_len=2, token_max_len=100, lower=True):
    lowerMap = {ord(u'A'): u'a',ord(u'A'): u'a',ord(u'B'): u'b',ord(u'C'): u'c',ord(u'Ç'): u'ç',ord(u'D'): u'd',ord(u'E'): u'e',ord(u'F'): u'f',ord(u'G'): u'g',ord(u'Ğ'): u'ğ',ord(u'H'): u'h',ord(u'I'): u'ı',ord(u'İ'): u'i',ord(u'J'): u'j',ord(u'K'): u'k',ord(u'L'): u'l',ord(u'M'): u'm',ord(u'N'): u'n',ord(u'O'): u'o',ord(u'Ö'): u'ö',ord(u'P'): u'p',ord(u'R'): u'r',ord(u'S'): u's',ord(u'Ş'): u'ş',ord(u'T'): u't',ord(u'U'): u'u',ord(u'Ü'): u'ü',ord(u'V'): u'v',ord(u'Y'): u'y',ord(u'Z'): u'z'}
    inp = inp.translate(lowerMap)
    tokenizer = TurkishTokenizer.DEFAULT
    tknL = tokenizer.tokenize(inp)
    print("in")
    return[
        utils.to_unicode(token.content) for token in tknL
        if(not token.content.startswith("_") and (token_min_len <= len(token.content) <= token_max_len))
    ]
```

```python
wiki = WikiCorpus('dump.xml.bz2')
print("Finished")
```

```python
lowerMap = {ord(u'A'): u'a',ord(u'A'): u'a',ord(u'B'): u'b',ord(u'C'): u'c',ord(u'Ç'): u'ç',ord(u'D'): u'd',ord(u'E'): u'e',ord(u'F'): u'f',ord(u'G'): u'g',ord(u'Ğ'): u'ğ',ord(u'H'): u'h',ord(u'I'): u'ı',ord(u'İ'): u'i',ord(u'J'): u'j',ord(u'K'): u'k',ord(u'L'): u'l',ord(u'M'): u'm',ord(u'N'): u'n',ord(u'O'): u'o',ord(u'Ö'): u'ö',ord(u'P'): u'p',ord(u'R'): u'r',ord(u'S'): u's',ord(u'Ş'): u'ş',ord(u'T'): u't',ord(u'U'): u'u',ord(u'Ü'): u'ü',ord(u'V'): u'v',ord(u'Y'): u'y',ord(u'Z'): u'z'}
txtF = open('dump.txt', 'w', encoding="utf-8")
ls = []
for part in wiki.get_texts():
    for subpart in part:
        subpart = subpart.translate(lowerMap)
        ls.append(subpart)
    txtF.write(utils.to_unicode(" ".join(ls.copy())+"\n"))
    ls.clear()
txtF.close()
```

```python
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence
```

```python
import multiprocessing
w2v = Word2Vec(LineSentence("dump.txt"), window=2, min_count=10 ,workers=multiprocessing.cpu_count())
```

```python
w2v.wv.save_word2vec_format("w2v_model1", binary=True)
```

```python
from gensim.models import KeyedVectors
wv1 = KeyedVectors.load_word2vec_format('w2v_model1', binary=True)
```

```python
wv1.most_similar(positive=["araba", "gemi"], negative=['uçak'])
```
