```python
from zemberek import (
    TurkishSpellChecker,
    TurkishSentenceNormalizer,
    TurkishSentenceExtractor,
    TurkishMorphology,
    TurkishTokenizer
)
from nltk import tokenize
from nltk.corpus import stopwords
```

```python
"""ds = open('C:\\Users\\disco\\Desktop\\ML\\dunya-nz.txt','r')
strK = ds.read()
ds.close()
print(strK)"""
strnk = "Yrn okua gidicem. Tmm, yarin havuza giricem ve aksama kadar yaticam :). ah aynen ya annemde fark ettı siz evinizden cıkmayın diyo. gercek mı bu? Yuh! Artık unutulması bile beklenmiyo. Hayır hayat telaşm olmasa alacam buraları gökdelen dikicem. yok hocam kesınlıkle oyle birşey yok. herseyi soyle hayatında olmaması gerek bence boyle ınsanların falan baskı yapıyosa. email adresim zemberek_python@loodos.com. Kredi başvrusu yapmk istiyrum. Bankanizin hesp blgilerini ogrenmek istyorum."
##sntc = tokenize.sent_tokenize(strnk)
##print(sntc)
ext = TurkishSentenceExtractor()
sntc = ext.from_paragraph(strnk)
print(sntc)
```

```python
morph = TurkishMorphology.create_with_defaults()
norm = TurkishSentenceNormalizer(morph)
normsent = []
for sn in sntc:
    normsent.append(norm.normalize(sn))
print(normsent)
```

```python
stp = set(stopwords.words('turkish'))
```

```python
tokenizer = TurkishTokenizer.DEFAULT
tkn = []
lmm = []
sent = []
for sn in normsent:
    tkn.append(tokenizer.tokenize(sn))
for tknlist in tkn:
    for t in tknlist:
        print('Content==', t.content)
        print('Type==', t.type_.name)
        print('Start==', t.start)
        print('End==', t.end, end='\n\n')
        if(t.content not in stp):
            if(t.type_.name != 'Punctuation' and t.type_.name != 'Emoticon'):
                lmm.append([morph.analyze(t.content)])
    #lmm.append(sent.copy())
    #sent.clear()
```

```python
sentenceList = []
ls = []
for l in lmm:
    for rs in l:
        for r in rs:
            s = str(r)
            ls.append(s.split(':')[0][1:])
            break
    sentenceList.append(ls.copy())
    ls.clear()
print(sentenceList)
```
